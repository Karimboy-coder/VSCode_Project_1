```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read informations
df = pd.read_csv('netflix_titles.csv')

# Let's see the top of the table.
df.head()
```

```python
# View table columns and data types
print(df.info())

# Statistics (how many rows, how many average values, etc.))
print(df.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=df, palette='magma')
plt.title('Netflix: Movie and TV Show Distribution')
plt.xlabel('Type')
plt.ylabel('Number')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title("Top 10 Countries That Produced the Most Content")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['release_year'].value_counts().sort_index().plot(kind='line', color='red', linewidth=2)
plt.title("Growth in content production over the years")
plt.xlabel('Release year')
plt.grid(True)
plt.show()
```
